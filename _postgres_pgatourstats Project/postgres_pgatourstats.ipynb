```python
create_table1 = ("""CREATE TABLE IF NOT EXISTS scoring_avg(
    RANK INT,
    PLAYER_ID INT PRIMARY KEY, 
    PLAYER VARCHAR, 
    "SCORING AVG" INT, 
    "TOTAL STROKES" INT, 
    "TOTAL ROUNDS" INT
    )""")
    
cur.execute(create_table1)
conn.commit()
```

```python
create_table2 = ("""CREATE TABLE IF NOT EXISTS driving_distance(
    RANK INT, 
    PLAYER_ID INT, 
    PLAYER VARCHAR, 
    "DISTANCE AVG" INT, 
    "TOTAL DISTANCE" INT, 
    "TOTAL DRIVES" INT
    )""")

cur.execute(create_table2)
conn.commit()
```

```python
create_table3 = ("""CREATE TABLE IF NOT EXISTS driving_accuracy(
    RANK INT, 
    PLAYER_ID INT, 
    PLAYER VARCHAR, 
    "DRIVING ACCURACY" FLOAT, 
    "FAIRWAYS HIT" INT, 
    "POSSIBLE FAIRWAYS" INT
    )""")
    
cur.execute(create_table3)
conn.commit()
```

```python
try:
    cur.execute("""
    SELECT s.*, d.rank as distance_rank, d."DISTANCE AVG" ,
    da."DRIVING ACCURACY", da.rank as driving_accuracy_rank
    FROM scoring_avg s 
    JOIN driving_distance d 
    ON d.player_id = s.player_id
    JOIN driving_accuracy da
    ON s.player_id = da.player_id
    ORDER BY s.rank;
    """)
except psycopg2.Error as e:
    print("query error")
    print(e)

conn.commit()
```
